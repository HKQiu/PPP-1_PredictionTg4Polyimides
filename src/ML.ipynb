```python
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np 
import re 
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
from pandas.core.common import random_state
import keras
```

```python
DATASET_FILE ='/content/drive/My Drive/ML/ML dataset.csv'
```

```python
dataset = pd.read_csv(DATASET_FILE)
dataset.drop(columns='Smiles', inplace=True)  
dataset.head()
```

```python
targets = dataset.iloc[:,-1]
features = dataset.drop(['Tg'], axis=1)

Xtrain, Xtest, Ytrain, Ytest = train_test_split(features, targets, test_size=0.2)

shuffle = True
random_state = 0
scoring = "neg_mean_absolute_error"
scoring1 = 'r2'
n_jobs = 10
```

# SVR

```python
model = SVR(kernel='rbf', C=200, gamma='auto')

cv = KFold(n_splits=10, shuffle=shuffle, random_state=random_state)
cv_res = cross_val_score(model,
              features,
              targets,
              cv=cv,
              scoring=scoring,
              n_jobs = n_jobs)


cv_res1 = cross_val_score(model,
              features,
              targets,
              cv=cv,
              scoring=scoring1,
              n_jobs = n_jobs)

print("'loss':" + str(-cv_res.mean()) + "(+/-" + str(cv_res.std()) + ")")
print("'R2':" + str(cv_res1.mean()) + "(+/-" + str(cv_res1.std()) + ")")
```
